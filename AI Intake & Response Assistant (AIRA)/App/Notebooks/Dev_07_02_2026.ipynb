```python
import pandas as pd

# If pycountry isn't installed in your notebook environment, run:
# !pip install pycountry

import pycountry

def alpha2_to_official_name(alpha2: str) -> str:
    if pd.isna(alpha2):
        return ""
    code = str(alpha2).strip().upper()
    if not code:
        return ""

    c = pycountry.countries.get(alpha_2=code)
    if not c:
        return ""  # unknown code

    # Prefer official_name if present, else fall back to name
    return getattr(c, "official_name", None) or c.name


# ---- Update input/output paths ----
input_csv = r"D:\Business\AI Intake & Response Assistant (AIRA)\App\Notebooks\Airports_with_IATA.csv"
output_csv = "output_with_country.csv"

df = pd.read_csv(input_csv)

# Add new column
df["Official Country Name"] = df["Alpha-2 code"].apply(alpha2_to_official_name)

df.to_csv(output_csv, index=False)

print("Saved:", output_csv)
df.head()
```

```python
# !pip install rapidfuzz

import pandas as pd
from rapidfuzz import process, fuzz

# ---- Load CSV ----
df = pd.read_csv("output_with_country.csv")  

# ---- Only keep these columns in final output ----
KEEP_COLS = ["IATA", "Name", "City", "Official Country Name"]

# ---- Build search keys (City + Name helps) ----
df["_key"] = (
    df["City"].fillna("").astype(str).str.strip()
    + " | "
    + df["Name"].fillna("").astype(str).str.strip()
)

keys = df["_key"].tolist()

def search_city(query: str, top_n: int = 10) -> pd.DataFrame:
    q = str(query).strip()
    if not q:
        return pd.DataFrame(columns=["confidence_score"] + KEEP_COLS)

    matches = process.extract(
        q,
        keys,
        scorer=fuzz.WRatio,   # good for typos + partial
        limit=top_n
    )
    # matches: [(matched_text, score, index), ...]

    idx = [m[2] for m in matches]
    scores = [m[1] for m in matches]

    out = df.loc[idx, KEEP_COLS].copy()
    out.insert(0, "confidence_score", scores)

    # Ensure descending order by score
    out = out.sort_values("confidence_score", ascending=False).reset_index(drop=True)
    return out

# ---- Example ----
search_city("Java Island", top_n=10)
```

```python
def search_city_markdown(query: str, top_n: int = 10) -> str:
    out = search_city(query, top_n=top_n)  # uses the function from previous cell

    # drop confidence_score as you requested
    out = out.drop(columns=["confidence_score"], errors="ignore")

    # convert to markdown
    return out.to_markdown(index=False)

# Example
print(search_city_markdown("amsterdm", top_n=10))
```

```python
import pandas as pd
from rapidfuzz import process, fuzz
from typing import Sequence

def search_city_markdown(
    df: pd.DataFrame,
    query: str,
    top_n: int = 10,
    keep_cols: Sequence[str] = ("City", "Official Country Name"),
    search_cols: Sequence[str] = ("Official Country Name",),   # <-- search only by City by default
    sep: str = " | ",
) -> str:
    # --- validations ---
    missing_keep = [c for c in keep_cols if c not in df.columns]
    if missing_keep:
        raise ValueError(f"Missing required keep_cols in df: {missing_keep}")

    missing_search = [c for c in search_cols if c not in df.columns]
    if missing_search:
        raise ValueError(f"Missing required search_cols in df: {missing_search}")

    q = str(query).strip()
    if not q:
        return pd.DataFrame(columns=list(keep_cols)).to_markdown(index=False)

    n = int(top_n) if top_n is not None else 10
    if n <= 0:
        return pd.DataFrame(columns=list(keep_cols)).to_markdown(index=False)

    # --- build search keys---
    key_series = df[search_cols[0]].fillna("").astype(str).str.strip()
    for col in search_cols[1:]:
        key_series = key_series + sep + df[col].fillna("").astype(str).str.strip()
    keys = key_series.tolist()

    # --- fuzzy match ---
    matches = process.extract(q, keys, scorer=fuzz.WRatio, limit=n)  # [(text, score, idx), ...]
    if not matches:
        return pd.DataFrame(columns=list(keep_cols)).to_markdown(index=False)

    idx = [m[2] for m in matches]
    scores = [m[1] for m in matches]

    out = df.loc[idx, list(keep_cols)].copy()
    out.insert(0, "_score", scores)

    # sort by score desc, then drop score
    out = (
        out.sort_values("_score", ascending=False)
           .drop(columns=["_score"])
           .reset_index(drop=True)
    )

    return out.to_markdown(index=False)


# Example:
md = search_city_markdown(df, "India", top_n=10)
print(md)
```

```python
import pandas as pd
import tiktoken

# ---- Config ----
csv_path = "output_with_country.csv"
col_name = "Official Country Name"
model_name = "gpt-4o-mini"  # change if you want a different tokenizer

df = pd.read_csv(csv_path)

# distinct, cleaned values
distinct_vals = (
    df[col_name]
    .dropna()
    .astype(str)
    .map(str.strip)
)
distinct_vals = distinct_vals[distinct_vals != ""].drop_duplicates()

# combine with SPACE (instead of newline)
combined_text = " ".join(distinct_vals.tolist())

enc = tiktoken.encoding_for_model(model_name)
total_tokens = len(enc.encode(combined_text))

print("Distinct values:", len(distinct_vals))
print("Total tokens (combined distinct values):", total_tokens)
```
